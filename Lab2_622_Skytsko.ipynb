## Імпорт необхідних бібліотек

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Базуючись на прикладі із лекції №3 побудувати алгоритм лінійної регресії

```python
# Generate the regression data
X, y = make_regression(n_samples=40, n_features=1, n_informative=1, noise=8, random_state=8)

# Create the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Get coefficients (a and b)
a = model.coef_[0]
b = model.intercept_

# Draw the result
fig = plt.figure(figsize=(10, 6))
x = np.arange(-3, 4)
y_x = a * x + b

plt.plot(x, y_x, linewidth=2, c='r', label=f'y = {a:.2f}x + {b:.2f}')
plt.scatter(X, y)
plt.grid()
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend(prop={'size': 18})
plt.show()

# Predict the target value for x = 2
print(model.predict([[2]]))

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)

print(f"Середньоквадратична помилка (MSE): {mse:.2f}")
```

## Базуючись на прикладах із лекції №3 побудувати алгоритми наївного Байєса, дерева рішень, випадкового лісу та градієнтного бустинга для датасету ірисів

### Імпорт необхідних бібліотек

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

import xgboost
```

### Наївний алгоритм Байєса

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Encode string class (species names) values as integers
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
y_encoded = label_encoder.transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=0)

# Create the Naive Bayes classifier
model = GaussianNB()

# Train and evaluate the classifier
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
```

### Дерево рішень

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Encode string class (species names) values as integers
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
y_encoded = label_encoder.transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=0)

# Create the Decision Tree classifier
model = DecisionTreeClassifier(criterion = 'entropy', random_state = 13)

# Train and evaluate the classifier
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
```
